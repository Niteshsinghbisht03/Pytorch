# **Numpy array to Torch Tensor**

```python
X_test_tensor=torch.from_numpy(X_test)
X_train_tensor = torch.from_numpy(X_train)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

# **Defining a Model**

```python
class mySimpleNN():
  #initialize the weights and bias
  def __init__(self,X):
    self.w=torch.rand(X.shape[1],1,requires_grad=True,dtype=torch.double)
    self.b=torch.zeros(1,dtype=torch.double,requires_grad=True)

  #Feed Forward Training
  def forward_pass(self,X):
    return torch.sigmoid(torch.matmul(X,self.w)+self.b)

  #loss calculation
  def loss(self,y_pred,y):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred,epsilon,1-epsilon)

    return -(y_train_tensor * torch.log(y_pred) + (1-y_train_tensor)*torch.log(1-y_pred)).mean()


```

```python
class Model(nn.Module):
  def __init__(self,num_features):
    super().__init__()

    self.linear = nn.Linear(num_features,1)
    self.sigmoid = nn.Sigmoid()

  def forward(self,features):
    out = self.linear(features)
    out=self.sigmoid(out)
    return out
```

# **Important Parameters**

```python
#Learning Rate and number of Epochs
lr=0.1
epochs=300
```

# **Training Pipeline**

```python
#create Model
model = mySimpleNN(X_train_tensor)

for epochs in range(epochs):
  #forward Prediciton
  y_pred = model.forward_pass(X_train_tensor)

  #loss calculation
  loss= model.loss(y_pred,y_train_tensor)

  #backpropogation
  loss.backward()

  #weight updation
  with torch.no_grad():
    model.w -= lr * model.w.grad
    model.b -= lr * model.b.grad


  model.w.grad.zero_()
  model.b.grad.zero_()

  print(f'epochs: {epochs+1 }  :  loss: {loss.item()}')
```

```python
#Pytorch Based Code
model2 = Model(X_train_tensor.shape[1])
```

```python
loss_function = nn.BCELoss()
```

```python
X_train_tensor = X_train_tensor.type(torch.float32)

optimizer = torch.optim.SGD(model2.parameters(),lr=lr)

for epochs in range(epochs):
  #forward Prediciton
  y_pred = model2(X_train_tensor)

  #loss calculation
  loss= loss_function(y_pred,y_train_tensor.view(-1,1))

  #clear gradient
  optimizer.zero_grad()

  #backpropogation
  loss.backward()

  #weight updation
  optimizer.step()


  print(f'epochs: {epochs+1 }  :  loss: {loss.item()}')

```
